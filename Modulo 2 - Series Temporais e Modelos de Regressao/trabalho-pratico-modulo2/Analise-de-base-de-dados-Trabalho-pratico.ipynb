## Sobre o problema

O uso de bicicletas como um meio de transporte ganhou muita força nos últimos anos, seja
por questões ambientais, de saúde ou até mesmo infraestrutura de trânsito. Para incentivar
o uso, cidades em todo o mundo têm implementado programas de compartilhamento de bicicleta. Nesses sistemas, as bicicletas são retiradas e devolvidas em quiosques automatizados espalhados em diversos pontos da cidade.


As plataformas de compartilhamento de bicicletas costumam coletar diversos tipos de
dados, entre eles: duração da viagem, localização inicial e final do percurso, entre outros.
Esses dados coletados pela plataforma, em conjunto com informações sobre o clima, trânsito e relevo, por exemplo, tornam possível uma análise mais robusta de
compartilhamento de bicicletas.

A seguir, um descritivo dos dados coletados:
- rec_id: índice do registro de locação.
- datetime: data.
- season: estação do ano (1: inverno, 2: primavera, 3: verão, 4: outono). Relativo ao
hemisfério norte.
- year: ano (0: 2011, 1:2012).
- month: mês (1 a 12).
- hour: hora do dia (0 a 23).
- is_holiday: booleano indicando feriado.
- weekday: dia da semana (0: domingo, 1: segunda-feira, ..., 6: sábado).
- is_workingday: booleano indicando dia útil.
‒ weather_condition: (1: limpo, 2: nublado, 3: chuva leve, 4: chuva forte).
- temp: temperatura escalada entre 0 e 1. Valor original em graus Celsius: -8 a 39.
- atemp: sensação térmica escalada entre 0 e 1. Valor original em graus Celsius: -16
a 50.
- humidity: umidade relativa (0 a 1).
- windspeed: velocidade do vento escalada entre 0 e 1 (máximo original: 67).
- casual: número de locações para usuários casuais.
- registered: número de locações para usuários registrados.
- total_count: contador total de aluguéis (casual+registered).

Esta atividade tem como objetivo analisar os dados coletados de compartilhamento de
bicicletas em uma cidade, coletados pela Universidade do Porto.

```python
# Informações sobre os anos
  ## 2011 - Locações de bicicletas efetuadas
ano0_2011 = bike_df[bike_df['year']==0]
locacoes2011 = ano0_2011['total_count'].sum()
print("Locações de bicicletas efetuadas em 2011:", locacoes2011)
print(ano0_2011.shape)

  ## 2012 - Locações de bicicletas efetuadas
ano0_2012 = bike_df[bike_df['year']==1]
locacoes2012 = ano0_2012['total_count'].sum()
print("Locações de bicicletas efetuadas em 2012:", locacoes2012)
print(ano0_2012.shape)
```

```python
# Informações sobre os anos
  ## 2011 - Número de registros
registros2011 = ano0_2011['registered'].sum()
print("Número de registros em 2011:", registros2011)

  ## 2012 - Número de registros
registros2012 = ano0_2012['registered'].sum()
print("Número de registros em 2012:", registros2012)
```

```python
# Estação do ano com a maior e com a menor média de locações de bicicletas
estacoes = bike_df.groupby('season')['total_count'].mean().sort_values(ascending=False)
```

```python
estacoes
# 1: inverno, 2: primavera, 3: verão, 4: outono
```

```python
# Horários do dia com a menor e maior média de locação de bicicletas
horarios = bike_df.groupby('hour')['total_count'].mean().sort_values(ascending=False)
```

```python
horarios 
# hora do dia (0 a 23)
```

```python
# Dia da semana com a maior e com a menor media de locações
dias_locacoes = bike_df.groupby('weekday')['total_count'].mean().sort_values(ascending=False)
```

```python
dias_locacoes
# dia da semana (0: domingo, 1: segunda-feira, ..., 6: sábado)
```

```python
# Horário de quarta-feira que contém a maior e a menor média de locações de bicicletas
quarta = bike_df[bike_df['weekday']==3]
quarta = quarta.groupby('hour')['total_count'].mean().sort_values(ascending=False)
quarta
```

```python
# Horário de sábado que contém a maior e a menor média de locações de bicicletas
sabado = bike_df[bike_df['weekday']==6]
sabado = sabado.groupby('hour')['total_count'].mean().sort_values(ascending=False)
sabado
```
